```
import numpy as np
f = lambda x: 1/(1 + np.exp(-x))
g = lambda x: 1.0/(1.0+x**2)
h = lambda x: np.sin(x)

x_observed = np.linspace(-10, 10, 11)
x_latent = np.linspace(-10, 10, 101)
```

```
%matplotlib inline
import matplotlib.pyplot as plt

for func in [f, g, h]:
    plt.scatter(x_observed, func(x_observed), label='observed')
    plt.plot(x_latent, func(x_latent), label='latent')
    plt.grid()
    plt.legend()
    plt.show()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

for n in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
    coefficients = np.polyfit(x_observed, f(x_observed), n)
    y_fit = np.poly1d(coefficients)
    plt.scatter(x_observed, f(x_observed), label='observed')
    plt.plot(x_latent, f(x_latent), label='latent')
    plt.plot(x_latent, y_fit(x_latent), label='formula ' + str(n))
    plt.grid()
    plt.legend()
    plt.show()
```
