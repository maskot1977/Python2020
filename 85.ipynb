```
url = 'https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/location.txt'
```

```
import urllib.request
urllib.request.urlretrieve(url, 'location.txt')
```

```
import pandas as pd

df = pd.read_csv('location.txt')
df
```

```
mat = df.iloc[:, 1:].values
mat
```

```
import numpy as np
from scipy.spatial import distance

dist_mat = distance.cdist(mat, mat, metric='euclidean') # ユークリッド距離
dist_mat
```

```
pd.DataFrame(
    dist_mat, 
    columns=df['Town'],
    index=df['Town']
    )
```
