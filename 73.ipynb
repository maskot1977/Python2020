```
import urllib.request 

url = "https://github.com/maskot1977/ipython_notebook/blob/master/toydata/Teikoku-Shoin-Japan.xlsx?raw=true"
urllib.request.urlretrieve(url, 'Teikoku-Shoin-Japan.xlsx') 
```

```
import pandas as pd
excel = pd.ExcelFile('Teikoku-Shoin-Japan.xlsx') 
```

```
for s in enumerate(excel.sheet_names):
    print(s) # シートの名前と番号を列挙
```

```
sheets = []
for sheet_index in [2, 77, 78, 83]:
    print(excel.sheet_names[sheet_index])
    sheet = excel.parse(excel.sheet_names[sheet_index], header=None)
    sheets.append(sheet)
```

```
sheets0 = sheets[0].iloc[2:49, 1:]
sheets0.columns = ['都道府県名', '人口密度　〔2018年〕']
sheets0
```

```
sheets1 = sheets[1].iloc[2:49, 1:]
sheets1.columns = ['都道府県名', '1人あたりの都市公園面積　〔2017年3月末〕']
sheets1
```

```
sheets2 = sheets[2].iloc[3:50, 1:]
sheets2.columns = ['都道府県名', '1人あたりの電力需要量　〔2015年〕']
sheets2
```

```
sheets3 = sheets[3].iloc[2:49, 1:]
sheets3.columns = ['都道府県名', '公害苦情件数　〔2017年〕']
sheets3
```

```
df = pd.merge(sheets0, sheets1, on='都道府県名')
df = pd.merge(df, sheets2, on='都道府県名')
df = pd.merge(df, sheets3, on='都道府県名')
df
```
