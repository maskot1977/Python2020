```
# 全n個の品物のうち、i番目以降の品物から、重さの総和が j 以下になるように選ぶ
def question42(i, j):
    if i == n:
        return 0
    elif j < W[i]:
        return question42(i + 1, j)
    else:
        return max(question42(i + 1, j), question42(i + 1, j - W[i]) + V[i])
```

```
def question42(i, j):
    if dp_table[i][j] >= 0:
        return dp_table[i][j]
    elif i == n:
        ret = 0
    elif j < W[i]:
        ret = question42(i + 1, j)
    else:
        ret = max(question42(i + 1, j), question42(i + 1, j - W[i]) + V[i])

    dp_table[i][j] = ret
    return ret
```
