```
import math
def rotation(x, t, deg = False):
    if deg == True:
        t = t * math.pi / 180
    a = [[math.cos(t), -math.sin(t)],
         [math.sin(t),  math.cos(t)]]
    ax = [x[0] * a[0][0] + x[1] * a[0][1], x[0] * a[1][0] + x[1] * a[1][1]]
    return ax
```

```
def draw2(x0, n):
    angle = 360 / n
    P = []
    P.append(x0)
    for i in range(n):
        x = rotation(P[i], angle * 2, deg=True)
        P.append(x)
    return P
```

```
%matplotlib inline
import matplotlib.pyplot as plt
```

```
x0 = [0, 1]
n = 5
P = draw2(x0, n)
plt.figure(figsize=(6,6))
plt.plot([x for x, y in P], [y for x, y in P])
plt.show()
```

```
def draw3(x0, n):
    angle = 360 / n
    P = []
    P.append(x0)
    for i in range(n):
        x = rotation(P[i], angle * int(n/2), deg=True)
        P.append(x)
    return P
```

```
x0 = [0, 1]
n = 19
P = draw3(x0, n)
plt.figure(figsize=(6,6))
plt.plot([x for x, y in P], [y for x, y in P])
plt.show()
```

```
def draw4(x0, n, m):
    angle = 360 / n
    P = []
    P.append(x0)
    for i in range(n):
        x = rotation(P[i], angle * m, deg=True)
        P.append(x)
    return P
```

```
x0 = [0, 1]
n = 19
plt.figure(figsize=(6,6))
for m in [2, 3, 4, 5, 6, 7, 8, 9, 10]:
    P = draw4(x0, n, m)
    plt.plot([x for x, y in P], [y for x, y in P])
plt.show()
```
