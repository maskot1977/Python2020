```
import urllib.request
import pandas as pd

# ニューヨークの大気状態観測値データ
url = 'https://raw.githubusercontent.com/maskot1977/ipython_notebook/master/toydata/airquality.txt' 
urllib.request.urlretrieve(url, 'airquality.txt')

df = pd.read_csv('airquality.txt', sep='\t', index_col=0) 
```

```
df.iloc[:, :4]
```

```
df.iloc[:, :4].apply(lambda x: (x-x.mean())/x.std(), axis=0)
```

```
df.iloc[:, :4].apply(lambda x: (x-x.mean())/x.std(), axis=1)
```

```
df.iloc[:, :4].apply(lambda x: (x-x.min())/(x.max() - x.min()), axis=0)
```

```
df.iloc[:, :4].apply(lambda x: (x-x.min())/(x.max() - x.min()), axis=1)
```

```
df.iloc[:, :4].apply(lambda x: x/x.sum(), axis=0)
```

```
df.iloc[:, :4].apply(lambda x: x/x.sum(), axis=1)
```
